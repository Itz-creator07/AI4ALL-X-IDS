## Section 2: Load CICIDS2017 Dataset

**Dataset Structure:**
- 8 CSV files (Monday-Friday traffic)
- ~2.8M total network flows
- 79 features + 1 label column
- 15 classes (1 BENIGN + 14 attack types)

```python
# List of CICIDS2017 files
network_files = [
    "Monday-WorkingHours.pcap_ISCX.csv",
    "Tuesday-WorkingHours.pcap_ISCX.csv",
    "Wednesday-workingHours.pcap_ISCX.csv",
    "Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv",
    "Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv",  # Note: typo in original filename
    "Friday-WorkingHours-Morning.pcap_ISCX.csv",
    "Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv",
    "Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv"
]

print("=" * 80)
print("LOADING CICIDS2017 NETWORK INTRUSION DETECTION DATASET")
print("=" * 80)
print(f"\nNumber of files to load: {len(network_files)}")
print("\nFiles:")
for i, f in enumerate(network_files, 1):
    print(f"  {i}. {f}")
```

```python
# Load and combine all CSV files
df_list = []
total_rows = 0
file_stats = []

print("\nüìÇ Loading files...\n")

for filename in tqdm(network_files, desc="Loading CICIDS2017 files"):
    filepath = os.path.join(DATA_PATH, filename)
    
    try:
        df = pd.read_csv(
            filepath,
            sep=',',
            on_bad_lines='skip',
            low_memory=False
        )
        
        # Clean column names (remove leading/trailing whitespace)
        df.columns = df.columns.str.strip()
        
        rows = len(df)
        total_rows += rows
        df_list.append(df)
        
        file_stats.append({
            'filename': filename,
            'rows': rows,
            'columns': len(df.columns)
        })
        
        print(f"  ‚úÖ {filename:65s} {rows:>10,} rows")
        
    except FileNotFoundError:
        print(f"  ‚ùå File not found: {filename}")
        print(f"     Expected at: {filepath}")
        print(f"     Please verify file location and update DATA_PATH if needed.")
    except Exception as e:
        print(f"  ‚ùå Error loading {filename}: {e}")

# Combine all dataframes
if len(df_list) > 0:
    print("\nüîÑ Combining all files...")
    df_network = pd.concat(df_list, ignore_index=True)
    del df_list  # Free memory
    
    print(f"\n{'='*80}")
    print("‚úÖ DATA LOADING COMPLETE")
    print(f"{'='*80}")
    print(f"   Total rows loaded: {total_rows:,}")
    print(f"   Combined shape: {df_network.shape}")
    print(f"   Total features: {df_network.shape[1] - 1} (+ 1 label column)")
    print(f"   Memory usage: {df_network.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
else:
    print("\n‚ùå ERROR: No files were successfully loaded!")
    print("   Please check DATA_PATH and file locations.")
```

## Section 4: Data Cleaning

**Critical Steps (in order):**
1. Replace inf/-inf with NaN
2. Drop rows with NaN
3. Normalize skewed features (log transform)
4. Verify no infinite values remain

```python
print("=" * 80)
print("DATA CLEANING")
print("=" * 80)

print(f"\nüìè Initial shape: {df_network.shape}")
rows_initial = len(df_network)

# Step 1: Replace infinite values with NaN
print("\n1Ô∏è‚É£ Handling infinite values...")
inf_count_before = np.isinf(df_network.select_dtypes(include=[np.number]).values).sum()
print(f"   Infinite values found: {inf_count_before:,}")

df_network.replace([np.inf, -np.inf], np.nan, inplace=True)

inf_count_after = np.isinf(df_network.select_dtypes(include=[np.number]).values).sum()
print(f"   Infinite values after replacement: {inf_count_after:,}")
print("   ‚úÖ Infinite values replaced with NaN")

# Step 2: Drop rows with missing values
print("\n2Ô∏è‚É£ Removing missing values...")
nan_count = df_network.isnull().sum().sum()
print(f"   NaN values to remove: {nan_count:,}")

df_network.dropna(inplace=True)

rows_after_cleaning = len(df_network)
rows_dropped = rows_initial - rows_after_cleaning
print(f"   Rows before: {rows_initial:,}")
print(f"   Rows after: {rows_after_cleaning:,}")
print(f"   Rows dropped: {rows_dropped:,} ({rows_dropped/rows_initial*100:.2f}%)")
print("   ‚úÖ Missing values removed")

# Step 3: Normalize highly skewed features
print("\n3Ô∏è‚É£ Normalizing skewed features...")
if 'Flow Duration' in df_network.columns:
    # Check skewness
    skewness_before = df_network['Flow Duration'].skew()
    print(f"   Flow Duration skewness before: {skewness_before:.2f}")
    
    # Ensure non-negative values
    df_network['Flow Duration'] = df_network['Flow Duration'].clip(lower=0)
    
    # Log transformation
    df_network['Flow Duration'] = np.log1p(df_network['Flow Duration'].astype(float))
    
    skewness_after = df_network['Flow Duration'].skew()
    print(f"   Flow Duration skewness after log1p: {skewness_after:.2f}")
    print("   ‚úÖ Applied log1p transformation to Flow Duration")

# Step 4: Final verification - remove any remaining inf/NaN
print("\n4Ô∏è‚É£ Final verification...")
df_network.replace([np.inf, -np.inf], np.nan, inplace=True)
df_network.dropna(inplace=True)

rows_final = len(df_network)
print(f"   Final shape: {df_network.shape}")
print(f"   Data retained: {rows_final/rows_initial*100:.2f}%")

# Verify data quality
print(f"\n‚úÖ Data Quality Check:")
print(f"   Missing values: {df_network.isnull().sum().sum()}")
print(f"   Infinite values: {np.isinf(df_network.select_dtypes(include=[np.number]).values).sum()}")
print(f"   Ready for modeling: {'YES ‚úÖ' if df_network.isnull().sum().sum() == 0 else 'NO ‚ùå'}")

print(f"\n{'='*80}")
print("‚úÖ DATA CLEANING COMPLETE")
print(f"{'='*80}")
```

---
## Section 5: Label Encoding

Create two types of labels:
1. **Binary:** 0=Benign, 1=Attack
2. **Multi-class:** 0-14 for each attack type

```python
print("=" * 80)
print("LABEL ENCODING")
print("=" * 80)

if 'Label' in df_network.columns:
    # Binary classification
    print("\n1Ô∏è‚É£ Creating Binary Labels (Benign vs Attack)...")
    df_network['is_attack'] = (df_network['Label'] != 'BENIGN').astype(int)
    
    print(f"\n   Binary Classification:")
    print(f"   Benign (0): {(df_network['is_attack'] == 0).sum():,} ({(df_network['is_attack'] == 0).mean()*100:.2f}%)")
    print(f"   Attack (1): {(df_network['is_attack'] == 1).sum():,} ({(df_network['is_attack'] == 1).mean()*100:.2f}%)")
    print("   ‚úÖ Binary labels created")
    
    # Multi-class encoding
    print("\n2Ô∏è‚É£ Creating Multi-class Labels (15 attack types)...")
    le = LabelEncoder()
    df_network['label_encoded'] = le.fit_transform(df_network['Label'])
    
    # Save label mapping - CONVERT numpy int32 to Python int for JSON
    label_mapping = {str(label): int(code) for label, code in zip(le.classes_, le.transform(le.classes_))}
    
    print(f"\n   Multi-class Encoding:")
    print(f"   Total classes: {len(le.classes_)}")
    print(f"   Encoded as: 0-{len(le.classes_)-1}")
    
    print(f"\n   üìã Label Mapping:")
    for label, code in sorted(label_mapping.items(), key=lambda x: x[1]):
        count = (df_network['label_encoded'] == code).sum()
        pct = count / len(df_network) * 100
        print(f"      {code:2d}: {label:30s} ({count:>8,} samples, {pct:5.2f}%)")
    
    # Save mapping to file (now with proper Python ints)
    mapping_file = os.path.join(OUTPUT_DIR, 'label_mapping.json')
    with open(mapping_file, 'w') as f:
        json.dump(label_mapping, f, indent=2)
    print(f"\n   ‚úÖ Label mapping saved to: {mapping_file}")
    
else:
    print("\n‚ùå ERROR: 'Label' column not found!")
```

## Section 6: Feature Selection

Reduce from 79 features to top 20-30 based on correlation with attack label.

```python
print("=" * 80)
print("FEATURE SELECTION")
print("=" * 80)

# Get numeric columns only (exclude label columns)
numeric_cols = df_network.select_dtypes(include=[np.number]).columns.tolist()
exclude_cols = ['is_attack', 'label_encoded']
feature_cols = [col for col in numeric_cols if col not in exclude_cols]

print(f"\nüìä Feature Analysis:")
print(f"   Total numeric features: {len(feature_cols)}")

# Calculate correlation with binary attack label
print("\nüîç Analyzing feature importance via correlation...")

correlation = df_network[feature_cols + ['is_attack']].corr()['is_attack'].drop('is_attack')
correlation_abs = correlation.abs().sort_values(ascending=False)

print(f"\nüìä Top 20 Features Correlated with Attacks:")
print("="*80)
for i, (feat, corr) in enumerate(correlation_abs.head(20).items(), 1):
    print(f"   {i:2d}. {feat:50s} r={correlation[feat]:+.4f} (|r|={corr:.4f})")

# Select top features
# Strategy: Keep features with |correlation| > 0.1 OR top 20, whichever is larger
threshold = 0.1
top_features_by_threshold = correlation_abs[correlation_abs > threshold].index.tolist()
top_features_by_rank = correlation_abs.head(20).index.tolist()

# Take union
selected_features = list(set(top_features_by_threshold + top_features_by_rank))
selected_features = [f for f in correlation_abs.index if f in selected_features]  # Maintain order

print(f"\n‚úÖ Feature Selection Results:")
print(f"   Features with |r| > {threshold}: {len(top_features_by_threshold)}")
print(f"   Top 20 by rank: {len(top_features_by_rank)}")
print(f"   Final selected features: {len(selected_features)}")

# Save selected feature names
feature_file = os.path.join(OUTPUT_DIR, 'feature_names.txt')
with open(feature_file, 'w') as f:
    for feat in selected_features:
        f.write(f"{feat}\n")
print(f"   ‚úÖ Feature names saved to: {feature_file}")

# Create reduced dataset
print(f"\nüì¶ Creating reduced dataset...")
df_reduced = df_network[selected_features + ['Label', 'is_attack', 'label_encoded']].copy()

print(f"\nüìè Dataset Dimensions:")
print(f"   Original: {df_network.shape}")
print(f"   Reduced: {df_reduced.shape}")
print(f"   Features reduced: {df_network.shape[1] - len(selected_features) - 3}")
print(f"   Memory saved: {(df_network.memory_usage(deep=True).sum() - df_reduced.memory_usage(deep=True).sum()) / 1024**2:.1f} MB")

# Update main dataframe
df_network = df_reduced
del df_reduced

print(f"\n{'='*80}")
print("‚úÖ FEATURE SELECTION COMPLETE")
print(f"{'='*80}")
```

```python
# Visualize feature correlations (top 15)
top_15_features = selected_features[:15] + ['is_attack']
corr_matrix = df_network[top_15_features].corr()

plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Feature Correlation Matrix (Top 15 Features)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'correlation_heatmap.png'), dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Correlation heatmap saved as 'correlation_heatmap.png'")
```

## Section 7: Train/Test Split

**Strategy:**
- 80/20 split (train/test)
- Stratified sampling (maintain class distribution)
- Binary classification target

```python
print("=" * 80)
print("TRAIN/TEST SPLIT")
print("=" * 80)

# Prepare features and labels
X = df_network[selected_features].copy()
y_binary = df_network['is_attack'].copy()
y_multiclass = df_network['label_encoded'].copy()

print(f"\nüìä Dataset Summary:")
print(f"   Total samples: {len(X):,}")
print(f"   Total features: {X.shape[1]}")
print(f"   Binary labels (is_attack): {y_binary.nunique()} classes")
print(f"   Multi-class labels: {y_multiclass.nunique()} classes")

# Stratified split (based on binary labels)
print(f"\nüîÄ Performing stratified train/test split (80/20)...")
X_train, X_test, y_train_binary, y_test_binary = train_test_split(
    X, y_binary,
    test_size=0.2,
    stratify=y_binary,
    random_state=RANDOM_SEED
)

# Also split multi-class labels (using same indices)
y_train_multi = y_multiclass.loc[X_train.index]
y_test_multi = y_multiclass.loc[X_test.index]

print(f"\n‚úÖ Split Complete:")
print(f"\n   Training Set:")
print(f"      Samples: {len(X_train):,}")
print(f"      Benign: {(y_train_binary == 0).sum():,} ({(y_train_binary == 0).mean()*100:.2f}%)")
print(f"      Attack: {(y_train_binary == 1).sum():,} ({(y_train_binary == 1).mean()*100:.2f}%)")

print(f"\n   Test Set:")
print(f"      Samples: {len(X_test):,}")
print(f"      Benign: {(y_test_binary == 0).sum():,} ({(y_test_binary == 0).mean()*100:.2f}%)")
print(f"      Attack: {(y_test_binary == 1).sum():,} ({(y_test_binary == 1).mean()*100:.2f}%)")

# Verify stratification
train_attack_pct = (y_train_binary == 1).mean() * 100
test_attack_pct = (y_test_binary == 1).mean() * 100
print(f"\n   Stratification Check:")
print(f"      Train attack %: {train_attack_pct:.2f}%")
print(f"      Test attack %: {test_attack_pct:.2f}%")
print(f"      Difference: {abs(train_attack_pct - test_attack_pct):.2f}% {'‚úÖ' if abs(train_attack_pct - test_attack_pct) < 1.0 else '‚ö†Ô∏è'}")
```

## Section 8: Save Processed Data

Save train/test splits for model training notebook.

```python
print("=" * 80)
print("SAVING PROCESSED DATA")
print("=" * 80)

# Save feature sets
print("\nüíæ Saving feature sets...")
X_train.to_csv(os.path.join(OUTPUT_DIR, 'X_train.csv'), index=False)
X_test.to_csv(os.path.join(OUTPUT_DIR, 'X_test.csv'), index=False)
print(f"   ‚úÖ Saved X_train.csv ({X_train.shape})")
print(f"   ‚úÖ Saved X_test.csv ({X_test.shape})")

# Save binary labels
print("\nüíæ Saving binary labels...")
y_train_binary.to_csv(os.path.join(OUTPUT_DIR, 'y_train_binary.csv'), index=False, header=['is_attack'])
y_test_binary.to_csv(os.path.join(OUTPUT_DIR, 'y_test_binary.csv'), index=False, header=['is_attack'])
print(f"   ‚úÖ Saved y_train_binary.csv ({len(y_train_binary):,} labels)")
print(f"   ‚úÖ Saved y_test_binary.csv ({len(y_test_binary):,} labels)")

# Save multi-class labels
print("\nüíæ Saving multi-class labels...")
y_train_multi.to_csv(os.path.join(OUTPUT_DIR, 'y_train_multiclass.csv'), index=False, header=['label_encoded'])
y_test_multi.to_csv(os.path.join(OUTPUT_DIR, 'y_test_multiclass.csv'), index=False, header=['label_encoded'])
print(f"   ‚úÖ Saved y_train_multiclass.csv ({len(y_train_multi):,} labels)")
print(f"   ‚úÖ Saved y_test_multiclass.csv ({len(y_test_multi):,} labels)")

# Save summary statistics
print("\nüíæ Saving dataset summary...")
summary = {
    'dataset': 'CICIDS2017',
    'total_samples': len(df_network),
    'total_features': len(selected_features),
    'train_samples': len(X_train),
    'test_samples': len(X_test),
    'num_classes': y_multiclass.nunique(),
    'class_balance': {
        'benign_pct': float((y_binary == 0).mean() * 100),
        'attack_pct': float((y_binary == 1).mean() * 100)
    },
    'random_seed': RANDOM_SEED,
    'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
}

summary_file = os.path.join(OUTPUT_DIR, 'dataset_summary.json')
with open(summary_file, 'w') as f:
    json.dump(summary, f, indent=2)
print(f"   ‚úÖ Saved dataset_summary.json")

print(f"\n{'='*80}")
print("‚úÖ ALL DATA SAVED SUCCESSFULLY")
print(f"{'='*80}")
print(f"\nüì¶ Output Files Created in '{OUTPUT_DIR}':")
print(f"   1. X_train.csv - Training features")
print(f"   2. X_test.csv - Test features")
print(f"   3. y_train_binary.csv - Training binary labels")
print(f"   4. y_test_binary.csv - Test binary labels")
print(f"   5. y_train_multiclass.csv - Training multi-class labels")
print(f"   6. y_test_multiclass.csv - Test multi-class labels")
print(f"   7. feature_names.txt - Selected feature list")
print(f"   8. label_mapping.json - Attack type encoding")
print(f"   9. dataset_summary.json - Dataset statistics")
print(f"  10. class_distribution.png - Visualization")
print(f"  11. correlation_heatmap.png - Visualization")
```
